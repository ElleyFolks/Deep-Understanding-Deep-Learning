# Data Exploration - Wine Quality Dataset

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

import numpy as np
import scipy.stats as stats

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

## Importing and Processing Data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

data = pd.read_csv(url, sep=';')
data
```

```python
# gives a statistical summary of the data
data.describe()
```

#### Note:
- For some of the features, they are on completely different scales. This dataset is going to have to normalized before having the model learn from it.
    - Example: Values for acidity are around 0.5, while values for fixed acidity are around 8.

```python
# listing number of unique values per column
for i in data.keys():
    print(f'{i} has {len(np.unique(data[i]))} unique values')
```

```python
# pairwise plot of the data
columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'quality']
sns.pairplot(data[columns], kind='reg', hue='quality')
plt.show()
```

```python
# plotting data
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.tick_params(axis='x', rotation=45)
plt.show()
```

### Note - Outliers are not good for deep learning models
- There are some points in the `total sulfur dioxide` feature that are greatly larger than the rest of the data points.
- If these points are determined to be outliers, they should be excluded from the cleaned dataset.
- In this example, these points are removed without checking to see if they're actually outliers.


```python
# removing outliers
data = data.drop(data[(data['total sulfur dioxide']>200)].index) # finds indices of rows where total sulfur dioxide > 200 and drops them

# alternate method, applying mask to data
#data = data[data['total sulfur dioxide'] < 200]

# plotting data before Z-score normalization
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.tick_params(axis='x', rotation=45)
plt.show()
```

### Z-score Normalization
$\dfrac{(x-mean)}{std\space dev}$, where the `x` is a data point, the `mean` is the average of all of the values in a column, and `std dev` is the standard deviation of all of the values in a column.
- This is used when values vary widely between data features (columns). 
- The values need to be in a similar range across features to ensure that the deep learning model can accurately identify relationships between them.
- The `quality` column should be not be normalized because it is the output that the model will try to predict.

```python
# Z-score normalization (except on quality feature)

# selecting columns to normalize
columns_score = data.keys()
columns_score = columns_score.drop('quality')

# Z-score normalization (manual calculation)
for column in columns_score:
    mean_value = np.mean(data[column])
    std_dev = np.std(data[column], ddof=1)
    data[column]= (data[column] - mean_value) / std_dev

# compact z-score normalization
# data[columns_score] = data[columns_score].apply(stats.zscore)

data.describe()
```

```python
# plotting normalized data
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data)
ax.tick_params(axis='x', rotation=45)
plt.show()
```

## Distribution of Quality Values
- In an ideal situation, all of the quality values are equally distributed through the entire dataset.
- Problems can arise when there are over-represented quality values.

```python
# plotting distribution of quality values
fig = plt.figure(figsize=(6,6))
plt.rcParams.update({'font.size': 22})

counts = data['quality'].value_counts()
plt.bar(list(counts.keys()), counts)
plt.xlabel('Wine Quality')
plt.ylabel('Count')
plt.title('Distribution of Wine Quality')
plt.show()
```

```python
# addressing the over-representation of quality 5 and 6

# creating a new column for boolean quality
data['bool_quality'] = 0

data['bool_quality'][data['quality'] >5] = 1 # Qualities 6, 7, 8 are considered good and are set to 1. Others are set to 0.

data[['quality', 'bool_quality']]
```

## Loading and Partitioning Data
- Converting arrays into Tensors
- loading data into Data Loaders
- Splitting Data into training and testing sets

```python
# converting to tensors
torch_data = torch.tensor(data[columns_score].values).float() # excluding quality and bool_quality columns
labels = torch.tensor(data['bool_quality'].values).float()

print(torch_data.shape)
print('labels array:',labels.shape)

# converting labels into a 1D tensor (from list)
labels = labels[:, None]
print('labels tensor:',labels.shape)
```

```python
# partitioning data into training and testing sets
training_data, test_data, training_labels, test_labels = train_test_split(torch_data, labels, test_size=0.1)

# creating tensor datasets
training_data = TensorDataset(training_data, training_labels)
test_data = TensorDataset(test_data, test_labels)

# translate into dataloaders
batch_size = 64
training_loader = DataLoader(training_data, batch_size=batch_size, shuffle=True, drop_last=True)
test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])
```
