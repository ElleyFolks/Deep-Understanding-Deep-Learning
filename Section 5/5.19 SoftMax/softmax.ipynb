# SoftMax function
## $\sigma_i = {e^{z_i}}/{\Sigma e^z}$
- **numerator** = ${e^{z_i}}$, takes $e^z$ for every value in a list "z"
- **denominator** = ${\Sigma e^z}$, the *sum* of $e^z$ for every value in a list "z"
- **sigma** = $numerator / denominator$, **output** of softmax calculation
<br>
***
### key points:
- resulting graph of softmax is logarithmic
- this is a relatively simple yet powerful function
- if the **sigma outputs** are plotted on a *logarithmic scale*, then the *result appears to be linear*
- this is a **linear transformation**

```python
# example 1) step by step computation of softmax

z = [1,3,5] # list of numbers (output of DL model)

numerator = np.exp(z)
denominator = np.sum(np.exp(z))
sigma = numerator / denominator

print("The softmax of z is:\n", sigma)
print("The sum of z is:\n",np.sum(sigma))
```

## Softmax function using NumPy

```python
# example 2) softmax example with random numbers
zed = np.random.randint(-5, high=15, size=25)

numerator = np.exp(zed)
denominator = np.sum(numerator)
sigma = numerator / denominator

# ploting results of random list

plot.plot(zed,sigma,'ko')
plot.xlabel('Numbers (zed)')
plot.ylabel('Softmax result $sigma$')
#plot.yscale('log')
plot.title('$\sum\sigma$ = %g'%np.sum(sigma)) # checks if sum of softmax is 1

plot.show()
```

### notable syntax differences:
1. must *create an instance* of the **softmax activation class**
2. **cannot** directly pass a numpy array OR a list into the class
3. the **list** *must be converted* into a **torch.Tensor** data type
