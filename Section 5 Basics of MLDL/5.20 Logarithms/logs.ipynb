# Logarithms!
- in the example, will use the *linspace* function to generate a sample dataset where all of the points are equally spaced
- Demonstrates logarithm function, and how it transforms the dataset.

```python
# parameters of linspace function: (start_point, ending_end, number_of_points)
sample = np.linspace(0.001,1,300)

# taking the log of the data set
log_sample = np.log(sample)
```

```python
# plotting the log function
graph = plot.figure(figsize=(10,2))

# example of how to increase font size
plot.rcParams.update({'font.size':20})

# plot(x-axis, y-axis)
plot.plot(sample,log_sample) 
plot.xlabel('sample')
plot.ylabel('log(sample)')

# actually shows graph
plot.show()
```
